```python
# The code below produces a plot of the Hermite Poisson flow as described in the paper:
# "Visualising the flows of orthogonal polynomial expansions of the Riemann Xi-function".
# The program reads the file with pre-computed (2*n) coefficients and produces an implicit plot, 
# i.e. it visualises the flow of the real roots when r varies. 
# Version 1.0, 2 april 2024. 
```

```python
var("r","t");

#Generate a plot of all ordinates of the non-trivial zeros < 60 to overlay the flow graphs
var('y')
ord = []
with open("ordinateszetazeros.txt", "r") as f: #file is stored on GitHub
    for row in f:
        y=float(row)
        if y < 60:
            ord.append([y,1])

z=list_plot(ord, size=15, marker="o", color = "black")

#Import the pre-computed (2*n) coefficients.
RDF=RealBallField(1024)
global m1; m1 = []; CoeffSum = RDF(0)
with open("Hermitecoeff", "r") as f:
    for row in f:
        CoeffSum = CoeffSum + RDF(row)
        m1.append([RDF(row)])
print("The sum of all the coefficients is : %3.16f = Xi(0) \n" % CoeffSum)

# Produce the implicit plot (i.e. show how the real zeros travel when r varies)       
CCF=ComplexBallField(1024)
def Xi(t):
    f=CCF("0.5",0)+I*CCF(t)
    xi=f/2*(f-1)*CCF(pi).pow(-f/2)*(f/2).gamma()*f.zeta()
    return xi
        
def hyp(t,n):
    hyp=CCF(t*t).hypergeometric([-n/2],[CCF("0.5")])
    return hyp

def HermiteFlow(t,r):
    sum=CCF(0)
    for n in range(300):
        sum=sum+CCF(str(r),0).pow(n)*CCF(str(m1[n]))*hyp(t,n)
    return real_part(sum)

print("Ensure 20 digits accuracy at t = 60: %3.20f (should show 20 zeros)" % real_part(HermiteFlow(60,1)-Xi(60)))

a1=implicit_plot(lambda t,r:((RDF(HermiteFlow(t,r)))),(t,0,60),(r,0,1.2),color='black',fontsize=13,linestyle ="-",linewidth=1,axes_labels=['$t$','$r$'],title=r'Trajectories of the real roots of $X^{\mathcal{H}}_r(t)$',gridlines=True,plot_points=120)
show(a1 + z,aspect_ratio='automatic')
```

```python
#Generate a plot of all ordinates of the zeros < 60 to overlay the flow graphs
var('y')
ord = []
for y in range(9):
    ord.append([2*pi()*(y+1),1])
z=list_plot(ord, size=15, marker="o", color = "black")

#Import the pre-computed (2*n) coefficients.
RDF=RealBallField(1024)
global m1; m1 = []; CoeffSum = RDF(0)
with open("Hermitecoeffint", "r") as f:
    for row in f:
        CoeffSum = CoeffSum + RDF(row)
        m1.append([RDF(row)])
print("The sum of all the coefficients is : %3.16f = Xin(0) \n" % CoeffSum)

# Produce the implicit plot (i.e. show how the real zeros travel when r varies)       
CCF=ComplexBallField(1024)
def Xin(t):
    Xin=2*CCF(t/2).sin()/t
    return Xin
        
def hyp(t,n):
    hyp=CCF(t*t).hypergeometric([-n],[CCF("0.5")])
    return hyp

def HermiteFlow(t,r):
    sum=CCF(0)
    for n in range(150):
        sum=sum+CCF(str(r),0).pow(n)*CCF(str(m1[n]))*hyp(t,n)
    return real_part(sum)

print("Ensure 20 digits accuracy at t = 60: %3.20f (should show 20 zeros)" % real_part(HermiteFlow(60,1)-Xin(60)))

ai=implicit_plot(lambda t,r:((RDF(HermiteFlow(t,r)))),(t,0,60),(r,0,1.2),color='black',linestyle ="-",linewidth=1,axes_labels=['$t$','$r$'],title=r'Trajectories of the real roots of $X^{\mathcal{H}}_r(t)$',gridlines=True,plot_points=120)
show(ai + z,aspect_ratio='automatic')
```
