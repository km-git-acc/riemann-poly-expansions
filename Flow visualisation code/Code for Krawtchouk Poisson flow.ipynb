```python
# The code below produces a plot of the Krawtchouk Poisson flow as described in the paper:
# "Visualising the flows of orthogonal polynomial expansions of the Riemann Xi-function".
# The program reads the file with pre-computed coefficients and produces an implicit plot, 
# i.e. it visualises the flow of the real roots when r varies. 
# Version 1.0, 2 april 2024. 
```

```python
var("r","t");

#Generate a plot of all ordinates of the non-trivial zeros < 60 to overlay the flow graphs
var('y')
ord = []
with open("ordinateszetazeros.txt", "r") as f: #file is stored on GitHub
    for row in f:
        y=float(row)
        if y < 60:
            ord.append([y,1])

z=list_plot(ord, size=15, marker="o", color = "black")

RDF=ComplexBallField(1024)
global m1; m1 = []; CoeffSum = RDF(0)
with open("KrawtchoukCoeffp12", "r") as f:
    for row in f:
        CoeffSum = CoeffSum + RDF(row)
        m1.append([RDF(row)])
print("The sum of all the coefficients is : %3.16f = Xi(0) \n" % CoeffSum)

CCF=ComplexBallField(2024)
def Xi(t):
    f=CCF("0.5",0)+I*CCF(t)
    xi=f/2*(f-1)*CCF(pi).pow(-f/2)*(f/2).gamma()*f.zeta()
    return xi

def hyp(t,p,N,n):
    hyp=CCF(1/p).hypergeometric([-n,-t],[-N+CCF("10e-200")])
    return hyp

def KrawtchoukFlow(t,r):
    sum=CCF(0)
    for n in range(201):
        sum=sum+CCF(str(r),0).pow(n)*CCF(str(m1[n]))*hyp(CCF(t),CCF("0.5"),CCF(200),n)
    return real_part(sum)

print("Ensure 20 digits accuracy at t = 60: %3.20f (should show 20 zeros)" % real_part(KrawtchoukFlow(60,1)-Xi(60)))

a1=implicit_plot(lambda t,r:(RDF(KrawtchoukFlow(t,r))),(t,0,60),(r,0,1.1),color='black',fontsize=13,linestyle ="-",linewidth=1,axes_labels=['$t$','$r$'],title=r'Trajectories of the real roots of $X^{\mathcal{K}}_r(t)$',gridlines=True,plot_points=150)
show(a1 + z,aspect_ratio='automatic')
```
